# **Exploratory Analysis for Dataset for Traffic Management**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import geohash2
import os
from multiprocessing import Pool, cpu_count
import warnings
warnings.filterwarnings("ignore")
```

## **1) Data Preprocessing**

```python
data_dir = '/home/angps/Documents/GrabChallenge/Traffic Management/Data'
df = pd.read_csv(os.path.join(data_dir,'training.csv'))
```

```python
def clean_data(df):
    df['latitude'] = df['geohash6'].apply(lambda x: float(geohash2.decode_exactly(x)[0]))
    df['longitude'] = df['geohash6'].apply(lambda x: float(geohash2.decode_exactly(x)[1]))
    df['latitude_error'] = df['geohash6'].apply(lambda x: float(geohash2.decode_exactly(x)[2]))
    df['longitude_error'] = df['geohash6'].apply(lambda x: float(geohash2.decode_exactly(x)[3]))
    df['Hour'] = df['timestamp'].apply(lambda x: int(x.split(':')[0]))
    df['Minute'] = df['timestamp'].apply(lambda x: int(x.split(':')[1]))
    df['Period'] = ((df['day'] - 1) * 24 * 4) + (df['Hour'] * 4) + df['Minute'] // 15 
    return df
    
def parrellize_clean_data(df, func):
    num_cores = max(1, cpu_count() - 2)  # at least 1 core used, -2 to prevent machine crashing
    num_partitions = num_cores
    df_split = np.array_split(df, num_partitions)
    pool = Pool(num_cores)
    df = pd.concat(pool.map(func, df_split))
    pool.close()
    pool.join()
    return df
```

```python
cleaned_df = parrellize_clean_data(df, clean_data)
cleaned_df = cleaned_df.drop(['timestamp'], axis=1)
cleaned_df.to_csv(os.path.join(data_dir,'cleaned_training.csv'),
                  index=False)
print("Preprocessing Done")
```

**Distribution of Demand**

```python
sns.distplot(df['demand'])
```

**Distribution of demand by day**

```python
demand_by_day_mean = df.groupby(['day']).mean()['demand']
demand_by_day_min = df.groupby(['day']).min()['demand']
demand_by_day_max = df.groupby(['day']).max()['demand']
demand_by_day_median = df.groupby(['day']).median()['demand']
demand_by_day_mean.plot(label='mean')
demand_by_day_min.plot(label='min')
demand_by_day_max.plot(label='max')
demand_by_day_median.plot(label='median')
plt.ylabel('Demand')
plt.legend()
```

**Distribution of demand by hour**

```python
demand_by_hour_mean = df.groupby(['Hour']).mean()['demand']
demand_by_hour_min = df.groupby(['Hour']).min()['demand']
demand_by_hour_max = df.groupby(['Hour']).max()['demand']
demand_by_hour_median = df.groupby(['Hour']).median()['demand']
demand_by_hour_mean.plot(label='mean')
demand_by_hour_min.plot(label='min')
demand_by_hour_max.plot(label='max')
demand_by_hour_median.plot(label='median')
plt.ylabel('Demand')
plt.legend()
```

**Visualising Total Aggregated Demand in different locations**

```python
grouped_location_df = df.groupby(['latitude','longitude']).sum()
grouped_location_df=grouped_location_df.reset_index()
grouped_location_df.head()
```

```python
def plt_heat_map(latitudes, longitudes, demand, title):
    fig = plt.figure(figsize=(14,8))
    plt.scatter(latitudes, longitudes, c=demand, cmap='bwr')
    plt.colorbar()
    plt.title(title,fontsize=16, fontweight='bold')
    plt.xlabel('Latitude', fontsize=16)
    plt.ylabel('Longitude', fontsize=16)
    plt.show()
```

```python
latitude = grouped_location_df['latitude']
longitude = grouped_location_df['longitude']
demand = grouped_location_df['demand']
plt_heat_map(latitude, longitude, demand, 'Total Aggregated Demand by location')
```
